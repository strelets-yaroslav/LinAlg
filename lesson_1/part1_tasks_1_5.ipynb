__1.__ Исследовать на линейную зависимость:

$$f_{1}(x)=e^{x}, f_{2}(x)=1, f_{3}(x)=x+1, f_{4}(x)=x-e^{x}.$$


Вектор $f_{4}(x)=x-e^{x}$ можно выразить через остальные вектора следующим образом: $(-1)\cdot f_{1}(x) + (-1)\cdot f_{2}(x) + f_{3}(x) = -e^{x} - 1 + x + 1 = x - e^{x} = f_{4}(x)$. Из этого следует, что данные вектора линейно зависимы.

__2.__ Исследовать на линейную зависимость:
$$f_{1}(x)=2, f_{2}(x)=x, f_{3}(x)=x^{2}, f_{4}(x)=(x+1)^{2}$$

Вектор $f_{4}(x)=(x+1)^{2}$ можно выразить через остальные вектора следующим образом: $\frac{1}{2}\cdot f_{1}(x) + 2\cdot f_{2}(x) + f_{3}(x) = 1 + 2x + x^{2}= (x + 1)^{2} = f_{4}(x)$. Из этого следует, что данные вектора линейно зависимы.

__3.__ Найти координаты вектора $x = (2, 3, 5)\in \mathbb{R}^{3}$ в базисе $b_{1}=(0, 0, 10)$, $b_{2}=(2, 0, 0)$, $b_{3}=(0, 1, 0)$.


$$x=(2,3,5)=(2,0,0)+(0,3,0)+(0,0,5)=1\cdot(2,0,0)+3\cdot(0,1,0)+\frac{1}{2}\cdot(0,0,10)=\frac{1}{2}b_{1}+b_{2}+3b_{3}$$

Таким образом, координаты вектора $x$ в заданном базисе будут следующими: $x=(\frac{1}{2},1,3)$.

а) в базисе $1$, $x$, $x^{2}$;

б) в базисе $x^{2}$, $x-1$, $1$.

a) в базисе $1$, $x$, $x^{2}$ вектор имеет координаты $(2, -2, 3): 3x^{2} - 2x + 2 = 2\cdot(1) + (-2)\cdot(x) + 3\cdot(x^{2}))$.

б) в базисе $x^{2}$, $x-1$, $1$ вектор имеет координаты $(3, -2, 0): 3x^{2} - 2x + 2 = 3x^{2} - 2(x - 1) = 3\cdot(x^{2}) + (-2)\cdot(x-1) + 0\cdot(1)$

__5.__ Установить, является ли линейным подпространством:

а) совокупность всех векторов трехмерного пространства, у которых по крайней мере одна из первых двух координат равна нулю;
    
б) все векторы, являющиеся линейными комбинациями данных векторов $\{u_{1}, u_{2}, \ldots, u_{n}\}$.

a) определение "хотя бы одна из первых двух координат равна 0" означает, что вектора $x = (0, a, b)$ и $y = (c, 0, d)$ удовлетворяют ему. Проверим, выполняются ли условия для подпространства:

1. $x + y = (0 + c, a + 0, b + d) = (c, a, b+d)$ - видно, что получившийся вектор уже не удовлетворяет начальному определению совокупности, следовательно, данная совокупность не может являться линейным подпространством.
2. $\alpha\cdot x = (0, \alpha\cdot a, \alpha\cdot b)$ - получившийся вектор удовлетворяет определению совокупности, но 1е условие не выполнено.

Данная совокупность не является линейным подпространством.

б) возьмём две линейных комбинации: $x = (\lambda_1 u_{1}, ..., \lambda_n u_{n})$ и $y = (\mu_1 u_{1}, ..., \mu_n u_{n})$ и проверим, выполняются ли условия для подпространства:

1. $x + y = (\lambda_1 u_{1} + \mu_1 u_{1}, ..., \lambda_n u_{n} + \mu_n u_{n}) = ((\lambda_1 + \mu_1)u_{1}, ..., (\lambda_n + \mu_n)u_{n}) = (\eta_1 u_{1}, ..., \eta_n u_{n})$.
2. $\alpha\cdot x = (\alpha\cdot(\lambda_1 u_{1}), ..., \alpha\cdot(\lambda_n u_{n})) = ((\alpha\cdot\lambda_1)u_{1}, ..., (\alpha\cdot\lambda_n)u_{n}) = (\eta_1 u_{1}, ..., \eta_n u_{n})$.

Оба условия выполняются, следовательно, данная совокупность являются линейным подпространством.
